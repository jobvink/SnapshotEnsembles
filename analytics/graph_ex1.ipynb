```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# for i in {1..20}; do cp $i/RNN-elec-snapshot-10M-24T-400E.csv data/$i.csv; done;
snapshot_path = 'experiments/RNN-elec-10M-24T-400E/data/'
single_model_path = 'experiments/RNN-elec-snapshot-10M-24T-400E/data/'
```

```python
single_data = pd.DataFrame()
for i in range(1, 21):
    instance = pd.read_csv(f'{single_model_path}{i}.csv')
    single_data = pd.concat([single_data, instance], axis=1)
    
snapshot_data = pd.DataFrame()
for i in range(1, 21):
    instance = pd.read_csv(f'{snapshot_path}{i}.csv')
    snapshot_data = pd.concat([snapshot_data, instance], axis=1)

single_data = single_data.drop(['Unnamed: 0'], axis=1)\
    .rename(columns={'accuracy': 'accuracy (Single Model)', 'val_accuracy': 'validation accuracy (Single Model)'})
snapshot_data = snapshot_data.drop(['Unnamed: 0'], axis=1)\
    .rename(columns={'accuracy': 'accuracy (Snapshot)', 'val_accuracy': 'validation accuracy (Snapshot)'})

data = pd.concat([single_data, snapshot_data], axis=1)
```

```python
sns.lineplot(data=data[['validation accuracy (Single Model)', 'validation accuracy (Snapshot)']]) # , err_style="bars", ci=68 for bars
plt.xlabel('Batch')
plt.ylabel('Accuracy')
plt.show()
```
